```python
import pandas as pd
```

```python
borderdata = pd.read_csv('data/CleanedBorderdata_2008-2018.csv')
gdpdata = pd.read_csv('data/CleanedGdp2008-2018.csv')
borderdata['Date'] = pd.to_datetime(borderdata['Date']).dt.date
borderdata['year'] = pd.DatetimeIndex(borderdata['Date']).year
gdpdata
```

```python
borderdata.head()
borderdata = borderdata.rename(columns = {'Latitutde':'Latitude'})
borderdata = borderdata[['Port Name','State','Port Code','Date','Measure','Value','Latitude','Longitude','Border']]
```

```python
borderdata['year'] = pd.DatetimeIndex(borderdata['Date']).year
```

```python
borderdatayear = borderdata.groupby(['year']).sum()
borderdatayear = borderdatayear.reset_index()
borderdatayear
```

```python
gdpusa = gdpdata.loc[gdpdata['Country Name'] == 'United States']
gdpmexico = gdpdata.loc[gdpdata['Country Name'] == 'Mexico']
gdpcanada = gdpdata.loc[gdpdata['Country Name'] == 'Canada']
gdpusacanada = gdpusa.merge(gdpcanada, how = 'outer')
gdpNA = gdpusacanada.merge(gdpmexico, how = 'outer')
gdpNA
```

```python
data = {'year':['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018'],'gdpNA':['2.01463','-10.7509','10.7714','8.36072','7.65415','5.52635','8.12477','6.85881','5.58486','7.31366','6.96026']}
gdpofNA = pd.DataFrame(data)
gdpofNA
```

```python
borderdatayear['gdpNA'] = gdpofNA['gdpNA']
borderdatayear = borderdatayear.rename(columns = {'Latitutde':'Latitude'})
borderdatayear
borderdatayear = borderdatayear[['year','Port Code','Latitude','Longitude','gdpNA','Value']]
borderdatayear
borderdatayear['gdpNA'] = gdpofNA['gdpNA']
borderdatayear['gdpNA'] = borderdatayear.gdpNA.astype(float)
borderdatayear.dtypes
borderdatayear['gdpNA'] = borderdatayear['gdpNA'] * (1000000000000)
borderdatayear
```

```python
dataset = borderdatayear.values
dataset
```

```python
X = dataset[:,0:5]
Y = dataset[:,5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split dataset into test and train
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
X_test_scaled
```

```python
from tensorflow.keras.utils import to_categorical
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initializing Neural Network
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'relu', input_dim = 5))
# Adding the second hidden layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```python
classifier.summary()
```

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Fitting our model 
classifier.fit(X_train, y_train, batch_size = 5, nb_epoch = 100)
```

```python
model_loss, model_accuracy = classifier.evaluate(
    X_train, y_train, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```
